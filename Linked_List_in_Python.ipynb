```python
class Node:                  # creates a class for each Node of Linked List
  def __init__(self,d,n):    # Takes input of element or data of a Node and the Link relation or pointer
    self.data = d
    self.next = n
def createlist(a):           # Taking an array or list as input to convert it to linked list
  head = Node(a[0],None)     # declaring head as the first Node by making the class object
  tail = head                # declaring a temporary variable to keep track of Node pointers
  for i in range(1,len(a)):  # Iterating the array or list to convert all data or element to Linked List
    n = Node(a[i],None)      # Creating a Node by making a class object
    tail.next = n            # Connecting Nodes with its previous Node for making a Linked List
                          # Here, tail.next connects the current Node (n) with its previous Node (previous n)
    tail = tail.next         # connecting temporary variable with the latest Node so that we can connect future Nodes
  return head               # returning the starting Node or start of the Linked List

createlist([1,2,3])
```

```python
class Node:
  def __init__(self,d,n):
    self.element = d
    self.next = n
class LinkedList:
  def __init__(self,a):
    self.head = Node(a[0],None)   
    tail = self.head
    for i in range(1,len(a)):
      n = Node(a[i],None)
      tail.next = n
      tail = tail.next
    
  def countNode(self):
    self.count = 0
    temp = self.head   
    while temp != None:
      self.count += 1
      temp = temp.next
    return print(self.count)

  def printList(self):
    x = self.head
    while x!= None:
      print(x.element, end = '')
      if x.next != None:
            print(', ', end = '')
      x = x.next  
  def nodeAt(self,idx):
    count = 0 
    temp = self.head
    obj = None
    while temp!=None:
      if count == idx:
        obj = temp
        break
      count += 1  
      temp = temp.next
    if obj == None:
      print("Invalid index")
    return print(obj)    

  def get(self,idx):
    count = 0
    temp = self.head
    obj = 0
    while temp != None:
      if count == idx:
        obj = temp.element
        break
      count += 1
      temp = temp.next
    return print(obj)   

  def set(self,idx,dat):
    count  = 0
    temp = self.head
    obj = None
    
    while temp != None:
      if count == idx:
        obj = temp.element
        temp.element = dat
        break
      count += 1
      temp = temp.next
    return print(obj)    

  def indexOf(self, dat):
      temp = self.head
      count = 0
      while temp != None:
          if temp.element == dat:
            return count
          count += 1
          temp = temp.next
      return -1

  def contains(self, dat):
    temp = self.head
    while temp != None:
      if temp.element == dat:
        return True
        break
      temp = temp.next
    return False

  def copyList(self):
    temp = self.head
    newhead = None
    while temp != None:
        n2 = Node(temp.element, None)
        if newhead == None:
            newhead = n2
            tail = newhead
        else:
            tail.next = n2
            tail = n2
        temp = temp.next
    return newhead


  def reverseList(self):
    newhead = None
    current = self.head
    while(current):
      n2 = Node(current.element,None)
      n2.next = newhead
      newhead = n2
      current = current.next
    return newhead
    
  def insert(self, elem, idx):
    if idx < 0 or idx > self.countNode():
        raise Exception('Invalid Index')
    newNode = Node(elem, None)
    if idx == 0:
        newNode.next = self.head
        self.head = newNode
    else:
        pred = self.nodeAt(idx - 1)
        newNode.next = pred.next
        pred.next = newNode

  def remove(self, idx):
    if idx < 0 or idx >= self.countNode():
        return None
    removeNode = self.nodeAt(idx)
    if idx == 0:
        self.head = removeNode.next
    else:
        pred = self.nodeAt(idx - 1)
        pred.next = removeNode.next
    return removeNode.element
  
  def rotateLeft(self):
    new_Last = self.head
    self.head = new_Last.next
    old_Last = new_Last
    while old_Last.next != None:
        old_Last = old_Last.next
    old_Last.next = new_Last
    new_Last.next = None        
  
  def rotateRight(self):
    old_Last = self.head
    while old_Last.next != None:
        new_Last = old_Last
        old_Last = old_Last.next
    new_Last.next = None
    old_Last.next = self.head
    self.head = old_Last



# x = LinkedList([1,2,3])
# x.countnode()
# x.printlist()
# x.nodeAt(0)
# x.get(0)
# x.set(0,-1)
# x.indexOf(9)
# x.reverseList()
```
